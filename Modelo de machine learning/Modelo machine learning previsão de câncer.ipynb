**UCI Machine Learning Repository**

**Algoritmos usado**

**K-NN**

**Informações sobre o atributo:**

1) ID número

2) Diagnóstico (M = maligno, B = benigno)

3-32)

**Dez recursos de valor real são calculados para cada núcleo celular:**

a) raio (média das distâncias do centro aos pontos no perímetro)

b) textura (desvio padrão dos valores da escala de cinza)

c) perímetro

d) área

e) suavidade (variação local nos comprimentos dos raios)

f) compacidade (perímetro ^ 2 / área - 1,0)

g) concavidade (severidade das porções côncavas do contorno)

h) pontos côncavos (número de porções côncavas do contorno)

i) simetria

j) dimensão fractal ("aproximação do litoral" - 1)

A média, o erro padrão e o "pior" ou maior (média dos três
maiores valores) desses recursos foram calculados para cada imagem,
resultando em 30 recursos. Por exemplo, o campo 3 é o raio médio, o campo
13 é o raio SE, o campo 23 é o pior raio.

Todos os valores de recursos são recodificados com quatro dígitos significativos.

Valores de atributo ausentes: nenhum

Distribuição de classes: 357 benignos, 212 malignos

# Treino e teste

```python
X = dados.iloc[:, 1:]
y = dados.iloc[:, 0]
```

```python
from sklearn.model_selection import train_test_split
SEED = 123

x_train, x_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.3, 
                                                    random_state=SEED)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler

Standard_scaler = StandardScaler
Standard_scaler = Standard_scaler().fit(x_train)
scaler_trasf = Standard_scaler.transform(x_train)
scaler_trasf
```

```python
from sklearn.preprocessing import MinMaxScaler

MinMax = MinMaxScaler().fit(x_train)
MinMax = MinMax.transform(x_train)
MinMax
```

# K-NN

```python
from sklearn.neighbors import KNeighborsClassifier

modelo_knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
modelo_knn.fit(x_train, y_train)
modelo_knn_pred = modelo_knn.predict(x_test)
modelo_knn_score = modelo_knn.score(x_train, y_train)
print("Previsão do K-NN", modelo_knn_pred)
```

```python
print("Score K-NN: %.2f" % (modelo_knn_score * 100))
```

```python
#Acurácia
from sklearn.metrics import confusion_matrix, accuracy_score

acuracia = accuracy_score(y_test, modelo_knn_pred)
print("Acurácia: %.2f" % (acuracia))
```

# Matriz de confusão

```python
# Matriz de confusão

matriz = confusion_matrix(y_test, modelo_knn_pred)
matriz
```

```python
mx = sns.heatmap(matriz, annot = True, fmt = "g", label = 10);

mx.set_title("Previsão câncer de mama", fontsize = 20)
mx.set_xlabel("Câncer para maligno", fontsize = 15)
mx.set_ylabel("Cêncer para benigno", fontsize = 15)
```

**Resultado para exame de cêncer**

0 = Para benigno resultado 101 + 1 = 103

1 = Para maligno resultado 15 + 53 = 68

# Classificação report - Sklearn metrics 

```python
from sklearn.metrics import classification_report

repor = classification_report(y_test, modelo_knn_pred)
print("K-NN\n")
print(repor)
```
